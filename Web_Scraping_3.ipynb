```python
# 1. Web Scra[pping Assignment 3
#import libraries
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
from selenium.webdriver.common.by import By
import time
#Chrome Version 105.0.5195.127
#downloaded the Chrome webdriver and extract all and 

#*** Get the input from the user
UserInput = str(input("Enter The Product Name to be Searched"))

#*** Connect to the Driver
driver = webdriver.Chrome(r"C:\Users\mrimc\Downloads\chromedriver_win32\chromedriver.exe")

#***Opening Amazon page on auto|mated chrome browser
driver.get("https://www.amazon.in/")

Product=driver.find_element(By.ID,"twotabsearchtextbox")
Product.send_keys(UserInput)

#** Search Click Event
search=driver.find_element(By.ID,"nav-search-submit-button")
search.click()


time.sleep(5)



 
```

```python


#***Create Empty List for above Elements
Brand_Name=[]
Price =[]
Return_Exchange=[]
Expected_Delivery =[]
Availability =[]
Product_URL =[]

def get_scrapped_data():
    result_tag = driver.find_element(By.CSS_SELECTOR,'div.s-main-slot.s-result-list.s-search-results.sg-row')
    items = result_tag.find_elements(By.XPATH,'//div[@data-component-type="s-search-result"]')
    for item in items:
        title = item.find_element(By.TAG_NAME,'h2').text
        price = item.find_element(By.CSS_SELECTOR,'.a-price').text.replace("\n",".")
        availability = '-'
        return_exchange = '-'
        expected_delivery = '-'
        product_url = item.find_element(By.TAG_NAME,'a').get_attribute("href")
        try:
            availability = item.find_element(By.CSS_SELECTOR,'.a-size-base.a-color-price').text
        except:
            pass

        try:
            expected_delivery = item.find_element(By.CSS_SELECTOR,'.a-row.a-size-base.a-color-secondary.s-align-children-center').text.replace("\n", " -")
        except:
            pass
        Brand_Name.append(title)
        Price.append(price)
        Return_Exchange.append(return_exchange)
        Expected_Delivery.append(expected_delivery)
        Product_URL.append(product_url)
        Availability.append(availability)
        
for j in range(0,3):
    get_scrapped_data()
    time.sleep(5)
    driver.find_element(By.CSS_SELECTOR,'.s-pagination-item.s-pagination-next.s-pagination-button.s-pagination-separator').click()    
    time.sleep(5)
    
#driver.close()
print(len(Brand_Name),len(Price),len(Return_Exchange),len(Expected_Delivery),len(Availability),len(Product_URL))
df=pd.DataFrame({'Brand Name':Brand_Name,'Price':Price,'Return Exchange':Return_Exchange,'Expected Delivery':Expected_Delivery,'Availability':Availability,'ProductURL':Product_URL})
display(df)
# *** Save it as CSv
df.to_csv('WebScrapping3_Q2.csv', header=False, index=False)
    
```

```python
##10 Hostel
#hostel name, distance from city centre, ratings, total reviews, 
#overall reviews, privates from price, dorms from price, facilities and property description
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
from selenium.webdriver.common.by import By
import time
#Chrome Version 105.0.5195.127
#downloaded the Chrome webdriver and extract all and 

#*** Connect to the Driver
driver = webdriver.Chrome(r"C:\Users\mrimc\Downloads\chromedriver_win32\chromedriver.exe")
# or driver=webdriver.Chrome(r"chromedriver.exe")

#***Opening Naukri page on automated chrome browser
driver.get("https://www.hostelworld.com/")
#*** Entering Designation and Location

time.sleep(5)
driver.find_element(By.ID,"location-text-input-field").click()
time.sleep(3)
SearchInput = driver.find_element(By.ID,"search-input-field")

SearchInput.send_keys("London")
time.sleep(10)
driver.find_element(By.CSS_SELECTOR,'ul#predicted-search-results > li:nth-child(2)').click()
driver.find_element(By.ID,"search-button").click()

time.sleep(10)
Name=[]
Distance=[]
Rating=[]
Total_Reviews=[]
Facilities=[]
Description=[]
Private_Price=[]
Dorm_Price=[]

def get_scrapped_data():
    container_tag = driver.find_element(By.CLASS_NAME,"page-inner")
    items = driver.find_elements(By.CLASS_NAME,"property-card")
    for item in items:
        property_title_box = item.find_element(By.CLASS_NAME,"property")
        
        name=property_title_box.find_element(By.CLASS_NAME,"title").text
        dist_from_centre=property_title_box.find_element(By.CSS_SELECTOR,'div.title-row > div.subtitle.body-3 > a > span.description').text
        rating = 'No rating'
        try:
            rating = property_title_box.find_element(By.CSS_SELECTOR,'div.info > div.bottom-rating > a > div > div.score.big').text
        except:
            pass
        total_reviews = property_title_box.find_element(By.CSS_SELECTOR,'div.info > div.bottom-rating > a > div > div.summary > div.reviews').text
        facilities="-"
        try:
            facilities = property_title_box.find_element(By.CSS_SELECTOR,'div.title-row > div.facilities-label.facilities').text.replace("\n",", ")
        except:
            pass
        description = property_title_box.find_element(By.CSS_SELECTOR,'div.info > div.rating-factors.prop-card-tablet.rating-factors.small').text.replace("\n",", ")
        Name.append(name)
        Distance.append(dist_from_centre)
        Rating.append(rating)
        Total_Reviews.append(total_reviews)
        Facilities.append(facilities)
        Description.append(description)
    
    items = driver.find_elements(By.CLASS_NAME,"property-card")
    for item in items:
        prices_box = item.find_element(By.CLASS_NAME,'prices-col')
        private_price = prices_box.find_elements(By.CSS_SELECTOR,"a.prices > div:nth-child(1)").text
        dorm_price = prices_box.find_elements(By.CSS_SELECTOR,"a.prices > div:nth-child(2)").text
        Private_Price.append(private_price)
        Dorm_Price.append(dorm_price)
         

get_scrapped_data()
df = pd.DataFrame({'Name':Name, 'Distance':Distance, 'Rating':Rating, 'Total Reviews':Total_Reviews, 'Facilities':Facilities,
                  'Description': Description, 'Private Price': Private_Price, 'Dorm Price': Dorm_Price})
df
```
